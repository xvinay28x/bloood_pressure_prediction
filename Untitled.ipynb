```python
import pandas as pd
```

```python
df = pd.read_csv("data (1).csv")
```

```python
df.head()
```

```python
df.corr()
```

```python
df.isna().sum()
```

```python
df.drop(["Patient_Number","Genetic_Pedigree_Coefficient","Physical_activity","salt_content_in_the_diet","alcohol_consumption_per_day","Level_of_Stress","Pregnancy"],inplace = True, axis = 1)
```

```python
df.fillna(0,inplace = True)
```

```python
x = df[["Level_of_Hemoglobin","Age","BMI","Sex","Smoking","Chronic_kidney_disease","Adrenal_and_thyroid_disorders"]]
y = df["Blood_Pressure_Abnormality"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
```

```python
predicted = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicted)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test,predicted)
MSE
```

```python
df = pd.DataFrame({"Actual":y_test,"predict":predicted})
```
